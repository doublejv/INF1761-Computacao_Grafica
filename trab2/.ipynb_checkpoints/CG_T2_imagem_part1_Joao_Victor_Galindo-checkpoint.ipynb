```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Leia e exiba uma imagem RGB.

path = 'images/'
name = 'lenna.png'
fname = path + name
fname
```

```python
# 1. Leia e exiba uma imagem RGB.

img = plt.imread(fname)

plt.imshow(img)
plt.show()
```

```python
# 2. Ilustre com um grafico xy a variação dos canais red, green e blue na linha correspondente a metade da altura.

h, w = img.shape[:2]

vx = [0, 511]
vy = [h//2, h//2]

plt.imshow(img)
plt.plot(vx, vy, 'k')
plt.show()
```

```python
# 2. Ilustre com um grafico xy a variação dos canais red, green e blue na linha correspondente a metade da altura.

red = img[h//2, :, 0]
green = img[h//2, :, 1]
blue = img[h//2, :, 2]
alpha = img[h//2, :, 3]

x = np.linspace(0, w - 1, w)

plt.plot(x, red, 'r')
plt.plot(x, green, 'g')
plt.plot(x, blue, 'b')
plt.plot(x, alpha, 'k')
plt.show()
```

```python
# 3. Transforme a imagem de RGB para uma imagem de luminância, Y, 
#    de acordo com a luminancia dos fósforos do sistema sRGB, explicada na pagina Relative luminance e exiba.

def gammainv(u):
    return 25 * u / 323 if u < 0.04045 else pow((200. * u + 11) / 211, 12. / 5)

def lum(rgb):
    red = gammainv(rgb[0])
    green = gammainv(rgb[1])
    blue = gammainv(rgb[2])
    
    return 0.2126 * red + 0.7152 * green + 0.0722 * blue


Y = np.zeros((h, w), dtype = np.float)

for y in range(h):
    for x in range(w):
        Y[y, x] = lum(img[y, x, :])
```

```python
# 3. Transforme a imagem de RGB para uma imagem de luminância, Y, 
#    de acordo com a luminancia dos fósforos do sistema sRGB, explicada na pagina Relative luminance e exiba.

plt.imshow(Y, cmap = 'gray')
plt.show()
```

```python
# 4. Transforme a imagem de luminância, Y, em imagem de "lightness", L, do sistema Lab

def lightness(Y):
    d = 6 / 29
    
    Y1 = pow(Y, 1/3) if Y > pow(d, 3) else Y / 3 * pow(d, 2) + 4 / 29
    L = 1.16 * Y1 - 0.16
    
    return L
```

```python
# 4. Transforme a imagem de luminância, Y, em imagem de "lightness", L, do sistema Lab

L = np.zeros((h, w), dtype = np.float)

for y in range(h):
    for x in range(w):
        L[y, x] = lightness(Y[y, x])
        
plt.imshow(L, cmap = 'gray')
plt.show()
```

```python
Y_ = np.linspace(0, 1, 256)
L_ = [lightness(y) for y in Y_]

plt.plot(Y_, L_)
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.show()
```
