```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path = "data_files/"
cie_file = path + "all_1nm_data.xls"
macbeth_file = path + "ColorChecker_RGB_and_spectra.xls"
planilha = "spectral_data"
print(cie_file, "\n", macbeth_file)
```

```python
cie = pd.read_excel(cie_file, header = 3)
cie
```

```python
# passando do "dat frame" para listas

nm = cie["nm"]
Aw = cie["CIE A"]
D65 = cie["CIE D65"]
V = cie["VM(l)"]
xbar = cie["x bar"]
ybar = cie["y bar"]
zbar = cie["z bar"]
```

```python
# passando para numpy array

Aw = np.asarray(Aw)
V = np.asarray(V)
D65 = np.asarray(D65)
xbar = np.asarray(xbar)
ybar = np.asarray(ybar)
zbar = np.asarray(zbar)
```

```python
# selecionar a parte que interessa (lambda de 380 a 780)

nm = nm[80:481]
Aw = Aw[80:481]
D65 = D65[80:481]
V = V[80:481]
xbar = xbar[80:481]
ybar = ybar[80:481]
zbar = zbar[80:481]
```

```python
Aw = Aw/np.amax(Aw)
D65 = D65/np.amax(D65)
```

```python
fig = plt.figure(figsize = (8, 8))

plt.plot(nm, Aw, 'red')
plt.plot(nm, D65, 'blue')
plt.plot(nm, V, 'black')
plt.plot(nm, V * Aw, 'green')
plt.show()
```
